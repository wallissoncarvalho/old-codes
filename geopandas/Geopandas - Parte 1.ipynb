Caso queiramos selecionar apenas apenas alguns polígonos do shapefile, podemos selecionar por atributos.

```python
#Selecionando apenas 2 municíos específicos
rj2=rj[rj['NM_MUNICIP'].isin(['RIO DE JANEIRO','APERIBÉ'])]
rj2.plot()
```

```python
#Selecionando todos, exceto os 2 municípios
rj3=rj[rj['NM_MUNICIP'].isin(['RIO DE JANEIRO','APERIBÉ'])==False]
rj3.plot()
```

```python
#Criando um shapefile apenas com o município do Rio de Janeiro e salvando
rj_municipio=rj[rj['NM_MUNICIP'].isin(['RIO DE JANEIRO'])]
rj.to_file('Dados\\RJ-MUNIP\\rj-municipio.shp')
```
